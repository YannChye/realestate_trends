```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
import numpy as np
```

```python
listings=pd.read_csv("../source/clean/listings_all_clean.csv")
```

```python
listings=listings.drop(["AdName","Feature","Area","Region","Postcode","Floorplan"],axis=1)
listings=pd.get_dummies(listings, columns=["SoldMethod"])
listings["SoldDate"]=pd.to_datetime(listings["SoldDate"],infer_datetime_format=True)  
listings["SoldYear"]=listings["SoldDate"].dt.year
listings.columns
```

```python
listings=pd.DataFrame(listings.groupby(["Suburb","lga","SoldYear","Type"]).agg(
    {'Id':'count',
     'Bath':'mean',
     'Bed':'mean',
     'Car':'mean',
     'Latitude':'mean',
     'Longitude':'mean',
     'SoldPrice':'mean',
     'offence_count':'mean',
     'crime_rate_lga':'mean',
     'SoldMethod_SoldByAuction':'sum',
     'SoldMethod_SoldByPrivateTreaty':'sum',
     'SoldMethod_SoldPriorToAuction':'sum'
    }))
listings=listings.reset_index(drop=False)
listings["Bath"]=listings["Bath"].astype(int)
listings["Bed"]=listings["Bed"].astype(int)
listings["Car"]=listings["Car"].astype(int)
listings["SoldPrice"]=listings["SoldPrice"].astype(int)
listings["offence_count"]=listings["offence_count"].astype(int)
listings["crime_rate_lga"]=listings["crime_rate_lga"].astype(int)
listings.head()
```

```python
listings.Type.unique()
```

```python
building_type=pd.DataFrame(
    {"type":listings.Type.unique()})
building_type
```

```python
building_type.to_csv("../database/building.csv",index=False)
```

```python
listings["Suburb"]=[x.lower() for x in listings["Suburb"]]
```

```python
suburb=listings[["Suburb","lga"]]
suburb=suburb.drop_duplicates(subset=['Suburb'],keep='first')
suburb.columns=['suburb','lga']
suburb.head()
```

```python
suburb.to_csv("../database/suburb.csv",index=False)
```

```python
year=pd.DataFrame({"year":listings["SoldYear"].unique()})
year.head()
```

```python
year.to_csv("../database/year.csv",index=False)
```

```python
offence_rate=pd.DataFrame(listings.groupby(["SoldYear","lga"])["crime_rate_lga"].mean())
offence_rate=offence_rate.reset_index()
offence_rate=offence_rate.rename(columns={"SoldYear":"year","crime_rate_lga":"crime_rate"})
offence_rate.head()
```

```python
offence_rate.to_csv("../database/offence_rate.csv",index=False)
```
